```python
import pandas as pd

# Load the dataset
file_path = "myexcel - myexcel.csv (1).csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info(), df.head()
```

```python
import numpy as np

# Replace "Height" with random values between 150 and 180
df["Height"] = np.random.randint(150, 181, size=len(df))

# Confirm changes
df.head()


```

```python
# Calculate the count and percentage of employees in each team
team_distribution = df["Team"].value_counts()
team_percentage = (team_distribution / len(df)) * 100

# Create a DataFrame for better visualization
team_analysis = pd.DataFrame({"Count": team_distribution, "Percentage": team_percentage.round(2)})

# Display the results
team_analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=10, kde=True, color="skyblue")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Age Distribution Among Employees")
plt.show()

```

```python
# Find the most common age
most_common_age = df["Age"].mode()[0]
print(f"Predominant Age Group: {most_common_age}")

```

```python
# Calculate total salary expenditure by team
team_salary = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)
highest_salary_team = team_salary.idxmax()
highest_salary_team_value = team_salary.max()

# Calculate total salary expenditure by position
position_salary = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)
highest_salary_position = position_salary.idxmax()
highest_salary_position_value = position_salary.max()

# Print results
print(f"Team with highest salary expenditure: {highest_salary_team} (${highest_salary_team_value:,.2f})")
print(f"Position with highest salary expenditure: {highest_salary_position} (${highest_salary_position_value:,.2f})")


```

```python
# Scatter plot for correlation
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation Between Age and Salary")
plt.show()

```

```python
# Calculate correlation
correlation = df[["Age", "Salary"]].corr().iloc[0, 1]
print(f"Correlation between Age and Salary: {correlation:.3f}")
```
